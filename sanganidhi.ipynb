```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_excel("C:\Users\santh\OneDrive\Desktop\archive[1]\diabetes_risk_prediction_dataset.csv", sheet_name='diabetes_risk_prediction_datase)


# Gender Distribution (Bar Plot)
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Age Distribution (Histogram)
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot: Age vs. Test Result (using 'class' as the test result)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='class', hue='Gender', data=df)  # 'class' is the test result column
plt.title("Age vs. Test Result")
plt.xlabel("Age")
plt.ylabel("Test Result (Positive/Negative)")
plt.show()

# Symptoms Frequency Analysis (Pie Chart for a specific symptom)
# Choose a symptom column, e.g., 'Polyuria'
plt.figure(figsize=(8, 6))
symptom_counts = df['Polyuria'].value_counts()
plt.pie(symptom_counts, labels=symptom_counts.index, autopct='%1.1f%%')
plt.title("Polyuria Frequency")
plt.show()

# Correlation Matrix
plt.figure(figsize=(12, 10))
# Compute the correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gender Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('gender_distribution.png')

# Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')

# Symptom Frequencies
plt.figure(figsize=(10, 6))
symptom_counts = df['Polyuria'].value_counts()
plt.bar(symptom_counts.index, symptom_counts.values)
plt.title('Polyuria Frequency')
plt.xlabel('Polyuria')
plt.ylabel('Frequency')
plt.savefig('polyuria_frequency.png')

# Symptom Co-occurrence
plt.figure(figsize=(12, 8))
symptoms = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness']
corr = df[symptoms].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Symptom Co-occurrence')
plt.savefig('symptom_cooccurrence.png')

# Outcome Distribution
plt.figure(figsize=(10, 6))
outcome_counts = df['class'].value_counts()
plt.pie(outcome_counts, labels=outcome_counts.index, autopct='%1.1f%%')
plt.title('Outcome Distribution')
plt.savefig('outcome_distribution.png')

# Age and Outcome
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='class', data=df)
plt.title('Age and Outcome')
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.savefig('age_and_outcome.png')

# Gender and Outcome
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='class', data=df)
plt.title('Gender and Outcome')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('gender_and_outcome.png')

# Correlation Between Symptoms
plt.figure(figsize=(12, 8))
corr = df[symptoms].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Symptoms')
plt.savefig('correlation_between_symptoms.png')

# Age vs. Outcome
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='class', data=df)
plt.title('Age vs. Outcome')
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.savefig('age_vs_outcome.png')

# Gender vs. Outcome
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='class', data=df)
plt.title('Gender vs. Outcome')
plt.xlabel('Gender')
plt.ylabel('Outcome')
plt.savefig('gender_vs_outcome.png')

# Symptom Distribution by Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Polyuria', data=df)
plt.title('Symptom Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency of Polyuria')
plt.savefig('symptom_by_gender.png')

# Symptom Distribution by Age Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Polyuria', data=df)
plt.title('Symptom Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Polyuria')
plt.savefig('symptom_by_age_group.png')
```

```python
import pandas as pd
import numpy as np

# Convert symptoms to numeric where 'Yes' = 1 and 'No' = 0
df['Polyuria'] = df['Polyuria'].map({'Yes': 1, 'No': 0})
df['Polydipsia'] = df['Polydipsia'].map({'Yes': 1, 'No': 0})
df['sudden weight loss'] = df['sudden weight loss'].map({'Yes': 1, 'No': 0})
df['weakness'] = df['weakness'].map({'Yes': 1, 'No': 0})

# Now try to generate the heatmap again
symptoms = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness']
corr = df[symptoms].corr()

# Plotting the heatmap
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Symptom Co-occurrence')
plt.savefig('symptom_cooccurrence.png')
```
